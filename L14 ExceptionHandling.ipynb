```python
# File Handling with Exception handling
# Write a program that opens a text file and write data to it as "Hello, Good Morning!!!". Handle exceptions that can be 
# generated during the I/O operations. Do not show the success message on the main exception handling block (write inside the 
# else block).

try:
    with open("greet.txt","r+") as f:
        f.write("Hello, Good Morning!!!")
except FileNotFoundError:
    print("File not found.")
else:
    print("Successfully written on the file.")
```

```python
# Number game program.
# Write a number game program. Ask the user to enter a number. If the number is greater than number to be guessed, raise a 
# ValueTooLarge exception. If the value is smaller the number to be guessed the, raise a ValueTooSmall exception and prompt the 
# user to enter again. Quit the program only when the user enters the correct number. Also raise GuessError if user guess a 
# number less than 1.

import random
result = random.randint(1,100)

class ValueTooLargeException(Exception):
    def __init__(self,message):
        print(message)
        
class ValueTooSmallException(Exception):
    def __init__(self,message):
        print(message)
        
class GuessError(Exception):
     def __init__(self,message):
        print(message)
        
while(True):
    try:
        guess = int(input(("Guess number between 1-100 : ")))
        if(guess>result):
            raise ValueTooLargeException("Guess smaller number") 
        elif(guess<result):
            raise ValueTooSmallException("Guess larger number") 
        elif(guess<1):
            raise GuessError("Guess number greater than 1") 
        else:
            print("You guess the answer")
            break
    except (ValueTooLargeException,ValueTooSmallException,GuessError) as e:
        pass
```

```python
# Cast vote
# Write a program that validate name and age as entered by the user to determine whether the person can cast vote or not. To 
# handle the age, create InvalidAge exception and for name, create InvalidName exception. The name will be invalid when the 
# string will be empty or name has only one word.
# Input:# Enter the name:               goransh singh
# Enter the age: 25
# Output:
# Goransh Singh  Congratulation !!! You can vote.

class InvalidAgeException(Exception):
    def __init__(self,message):
        print(message)
        
class InvalidNameException(Exception):
    def __init__(self,message):
        print(message)
        
try:
    age = int(input("Enter age : "))
    name = input("Enter name : ")
    if(age<18 or age>120):
        raise InvalidAgeException("Either your age is less than 18 or your age is more than 120")
    if(name == ' ' or len(name.split(' '))<2):
        raise InvalidNameException("Either your name entry is blank or you don't enter full name.")
except (InvalidAgeException,InvalidNameException) as e:
    pass
else:
    print("{} Congratulation !!! You can vote.".format(name))
finally:
    print("Thank u")
```

```python
# Write a python function which infinitely prints natural numbers in a single line. Raise the StopIteration exception after 
# displaying first 20 numbers to exit from the program.

n = 1
while(True):
    try:
        print(n,end=' ')
        n = n + 1
        if(n>20):
            raise StopIteration
    except StopIteration:
        print("\nStop Iteration")
        break
```
