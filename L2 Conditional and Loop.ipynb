```python
# Write a menu-driven program -
# cm to ft
# km to miles
# USD to INR
# exit

print('''
Hi select anb option 
1. for converting cm to ft
2. for km to miles
3. for USD to INR
4. for exit
''')
while(True):
    option = int(input("Choose option : "))
    if(option == 1):
        cm = float(input("Length(cm) : "))
        ft = cm * 0.0328084
        print("Length(ft.): ",round(ft,2))
    elif(option == 2):
        km = float(input("Distance(km) : "))
        mile = km * 0.621371
        print("Distance(mile) : ",round(mile,2))
    elif(option == 3):
        usd = float(input("Currency($) : "))
        rupee = usd * 81.67
        print("Currency(Rs) : ",round(rupee,2))
    elif(option == 4):
        break
    else:
        print("Wrong Input")
```

```python
# Display Fibonacci series up to 10 terms.
# Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The 
# first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34

n1 = 0
n2 = 1
print(n1,end='\t')
print(n2,end='\t')
for i in range(10):
    n3 = n1 + n2
    n1 = n2
    n2 = n3
    print(n3,end ='\t')
```

```python
# Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that 
# number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't 
# use for loop to solve this problem.

N = int(input("Number : "))
i = 1
s = 0
while(i != N+1):
    if(i%5==0):
        i = i + 1
        continue
    s = s + i
    if(s > 300):
        s = s - i
        break
    i = i + 1
print("Sum : ",s)
```

```python
# Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of 
# the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.

for i in range(1000,3000):
    flag = True
    num = i
    while(i>0):
        rem = i % 10
        if(rem % 2 != 0):
            flag = False
            break
        i = i // 10
    if(flag == True):
        print(num,end='\t')
```

```python
# A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and 
# RIGHT with a given steps.
# The trace of robot movement is shown as the following:
# UP 5
# DOWN 3
# LEFT 3
# RIGHT 2
# !
# The numbers after the direction are steps.
# ! means robot stop there.
# Please write a program to compute the distance from current position after a sequence of movement and original point.
# If the distance is a float, then just print the nearest integer.

pos = [0,0]
while(True):
    s = input('Enter a robot path : ').upper()
    if(s=='!'):
        break
    direction = s.split()[0]
    steps = int(s.split()[1])
    if(direction=='UP'):
        pos[1] = pos[1] + steps
    elif(direction == 'DOWN'):
        pos[1] = pos[1] - steps
    elif(direction=='LEFT'):
        pos[0] = pos[0] - steps
    elif(direction=='RIGHT'):
        pos[0] = pos[0] + steps
    else:
        pass
distance = (pos[0]**2+pos[1]**2)**0.5
print('Position : ',tuple(pos))
print("Distance : ",int(distance)+1)
```

```python
# Print all the Armstrong numbers in a given range.
# Range will be provided by the user
# Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the 
# Armstrong numbers.

l_limit  = int(input('Lower Limit : '))
u_limit = int(input('Upper Limit : '))
for i in range(l_limit,u_limit+1):
    num = i
    s  = 0
    while(num>0):
        s = s + (num % 10)**3
        num = num // 10
    if(s==i):
        print(i,end='\t')
```

```python
# Calculate the angle between the hour hand and minute hand.
# Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final 
# result angle. For example, if the final angle is 10.61, we need to print 10.
# Input:
# H = 9 , M = 0
# Output:
# 90
# Explanation:
# The minimum angle between hour and minute hand when the time is 9 is 90 degress.

from math import floor
hour = int(input('Hour : '))
minute = int(input('Minute : '))
angle = abs(30*hour - 5.5*minute)
angle = floor(min(360-angle,angle))
print("Angle between hour hand and minute hand is : ",angle)
```

```python
# Print the following pattern. Write a program to use for loop to print the following reverse number pattern.
# 5 4 3 2 1 
# 4 3 2 1 
# 3 2 1 
# 2 1 
# 1

rows = int(input("Rows : "))
for i in range(0,rows):
    for j in range(rows-i,0,-1):
        print(j,end='\t')
    print()
```

```python
# Print the following pattern.
# * 
# * * 
# * * * 
# * * * * 
# * * * * * 
# * * * * 
# * * * 
# * * 
# *

rows = int(input("Rows : "))
# k = 1
for i in range(0,rows+1):
    if(i<rows/2+1):
        print('*'*i,end='')
    else:
        print('*'*(rows-i+1),end='')
    print()
```

```python
# Write a program that will take a decimal number as input and prints out the binary equivalent of the number

decimal = int(input('Decimal Number : '))
binary = ''
while(decimal>0):
    rem = decimal % 2
    binary = str(rem) + binary
    decimal = decimal // 2
print('Binary Equivalent : ',binary)
```
