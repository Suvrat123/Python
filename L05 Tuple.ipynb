```python
# Join Tuples if similar initial element
# While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the 
# similarity of initial element. This problem can have applications in data domains such as Data Science.
# Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)] 
# Output : [(5, 6, 7, 8), (6, 10), (7, 13)] 

test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)] 
unique = []
result = []
for i in test_list:
    unique.append(i[0])
unique = set(unique)
for i in unique:
    result.append([i])
    for j in test_list:
        if(j[0]==i):
            result[-1].append(j[1])
result = list(map(tuple,result))
print('Output:\n',result)
```

```python
# Multiply Adjacent elements (both side) and take sum of right and left side multiplication result.
# The original tuple : (1, 5, 7, 8, 10)
# Resultant tuple after multiplication : 
# (1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)
# output-(5, 40, 91, 136, 80)

tup1 = (1, 5, 7, 8, 10)
result = []
for i in range(len(tup1)):
    if(i==0):
        result.append(tup1[i] * tup1[i+1])
    elif(i==len(tup1)-1):
        result.append(tup1[i-1] * tup1[i])
    else:
        result.append(tup1[i-1] * tup1[i] + tup1[i] * tup1[i+1])
print('Output:\n',tuple(result))
```

```python
# Check is tuples are same or not?
# Two tuples would be same if both tuples have same element at same index
# t1 = (1,2,3,0)
# t2 = (0,1,2,3)
# t1 and t2 are not same

t1 = (1,2,3,0)
t2 = (0,1,2,3)
if(t1==t2):
    print('Tuple are same')
else:
    print('Tuple are not same')
```

```python
# Count no of tuples, list and set from a list
# list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
# Output:
# List-2
# Set-2
# Tuples-1

list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]
result = [0,0,0]
for i in list1:
    if(type(i) == list):
        result[0] += 1
    elif(type(i) == set):
        result[1] += 1
    elif(type(i) == tuple):
        result[2] = 1
print(f'List-{result[0]}\nSet-{result[1]}\nTuple-{result[2]}')
```
